```python
import numpy as np
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
import os

df_seasons = pd.read_csv("f1data/seasons.csv")
```

```python
# Dictionary to store our dynamic variables
dynamic_vars = {}

# Read the CSV file
with open('f1data/seasons.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        if len(row) >= 2:  # Ensure the row has at least 2 columns
            year = row[0]
            url = row[1]

            # Create a variable name using the year and assign the URL to it
            var_name = f"year_{year}"
            dynamic_vars[var_name] = url
```

```python
# Create a directory to store the text files
if not os.path.exists('seasons_data'):
    os.makedirs('seasons_data')

# Read the CSV file and scrape each URL
with open('f1data/seasons.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        if len(row) >= 2:  # Ensure the row has at least 2 columns
            year = row[0]
            url = row[1]
            
            # Send GET request
            try:
                response = requests.get(url)
                response.raise_for_status()  # Raise an exception for bad status codes
            except requests.RequestException as e:
                print(f"Error fetching {url}: {e}")
                continue

            # Parse HTML
            soup = BeautifulSoup(response.text, 'html.parser')

            # Extract text
            text = soup.get_text()

            # Create a filename based on the year
            filename = f"seasons_data/year_{year}.txt"

            # Write to file
            try:
                with open(filename, "w", encoding="utf-8") as output_file:
                    output_file.write(text)
                print(f"Successfully scraped and saved data for year {year}")
            except IOError as e:
                print(f"Error writing to file for year {year}: {e}")

print("Scraping completed.")
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import os
import re

# Create a directory to store the text files
if not os.path.exists('race_data'):
    os.makedirs('race_data')

# Function to create a valid filename
def create_valid_filename(name, year):
    # Remove or replace invalid filename characters
    valid_name = re.sub(r'[\\/*?:"<>|]', "", name)
    return f"{valid_name}_{year}.txt"

# Read the CSV file and scrape each URL
with open('f1data/races.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        name = row['name']
        year = row['year']
        url = row['url']
        
        # Send GET request
        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise an exception for bad status codes
        except requests.RequestException as e:
            print(f"Error fetching {url}: {e}")
            continue

        # Parse HTML
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text
        text = soup.get_text()

        # Create a filename based on the name and year
        filename = create_valid_filename(name, year)
        filepath = os.path.join('race_data', filename)

        # Write to file
        try:
            with open(filepath, "w", encoding="utf-8") as output_file:
                output_file.write(text)
            print(f"Successfully scraped and saved data for {name} {year}")
        except IOError as e:
            print(f"Error writing to file for {name} {year}: {e}")

print("Scraping completed.")
```
